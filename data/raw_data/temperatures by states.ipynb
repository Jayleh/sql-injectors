```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
temp_df = pd.read_csv("GlobalLandTemperaturesByState_1998-2012.csv")
temp_df.head(20)
```

```python
temp_df.dtypes
```

```python
temp_df["year"] = temp_df["year"].values.astype(int)
temp_df.head(20)
```

```python
temp_max = temp_df.groupby(["State", "year"])["AverageTemperature"].max()
temp_max.head(20)
```

```python
temp_min = temp_df.groupby(["State", "year"])["AverageTemperature"].min()
temp_min.head(20)
```

```python
temp_max.to_csv("temp_max.csv")
temp_min.to_csv("temp_min.csv")
```
